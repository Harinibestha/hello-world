```python
## Groupby-to gain meaningful insights
#what is the average qty sold by reqion?
data.groupby('Region', as_index=False).agg({"Quantity":np.mean})
```

```python
data.groupby('Region', as_index=False).agg({"Quantity":"mean"})
```

```python
#what is the total sales by categories?
data.groupby('Category', as_index=False).agg({"Sales":np.sum})
```

```python
data.groupby('Segment')[['Sales','Profit']].agg({"Sales":np.sum,"Profit":np.max})
```

```python
data.groupby('Segment')[['Sales','Profit']].agg({"sum","max"})
```

```python
#what is the average discount given by segment?
data.groupby('Segment').agg({"Discount":np.mean})
```

```python
##Polishing the output and sorting
data.groupby('Segment',as_index=False).agg({"Discount":np.mean}).rename(columns={"Discount":"Average Discount"})
```

```python
data.groupby('Segment',as_index=False).agg({"Discount":np.mean}).rename(columns={"Discount":"Average Discount"}).sort_values("Average Discount",ascending=False)
```

```python
##which segment customers are most profitable
data.groupby('Segment',as_index=False).agg({"Profit":np.max}).sort_values("Profit",ascending=False)
```

```python
#which are the top 5 categories to give the max profit

data.groupby('Category').agg({"Profit":np.max}).sort_values("Profit",ascending=False)
```

```python
#Adding new columns
#comparing sales per order with average sales
data["Hi_Low"]=(data['Sales'])/(data['Sales'].mean())
data.head()
```

```python
data["Cost"]=(data['Sales'])-(data['Profit'])
data.head()
```

```python
##Apply
##what is the total sales,qty,discount,profit across total us?
def get_sum(x):
    return np.sum(x)
data[["Sales","Quantity","Discount","Profit"]].apply(get_sum,axis=0)
```

```python
##Datetime
#Find how much time it takes to place an order and ship the product
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Ship Date']=pd.to_datetime(data['Ship Date'])
```

```python
data['Duration']=data['Ship Date']-data['Order Date']
data.head()
```

```python
#How many days it has taken more than 5 days from placing an order to shipping
data[data['Duration']>'5 days'].shape[0]
```
